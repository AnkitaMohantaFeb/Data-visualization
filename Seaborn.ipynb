```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
print(sns.get_dataset_names())
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.set_style('whitegrid')
sns.histplot(data=titanic,x='fare',binwidth=20,color='#F08080')
plt.axis([0,600,0,500])
```

```python
#tips["tip"].plot(kind='box')

obj=sns.boxplot(data=titanic,x=titanic['class'],y=titanic['age'],
            palette='rainbow')
```

Create a bar chart for the number of male and female passengers.

```python
count=titanic["sex"].value_counts()
print(count)
```

```python
sns.set_style('whitegrid')
sns.countplot(data=titanic,x=titanic['sex'],palette=['#4682B4','#009E60'])
```

```python
sns.set_style('darkgrid')
g=sns.PairGrid(titanic,x_vars=['class','sex','who','alone'],y_vars='survived',height=4,aspect=0.5)
g.map(sns.pointplot,color='#630330')
g.set(ylim=(0,1))
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
sns.barplot(data=diamonds,x='cut',y='price',hue='color',
            hue_order=['E', 'I', 'J', 'H', 'F', 'G', 'D'],
            order=['Ideal','Premium','Good','Very Good','Fair'])
plt.legend(ncol=4)
plt.xlabel('Cut',fontsize=14)
plt.ylabel('Price',fontsize=14)
plt.show()
```

```python
sns.set_style('white')
sns.countplot(data=diamonds,x='cut',color='red',
              order=['Ideal','Premium','Good','Very Good','Fair'])
sns.despine(top=True,right=True,left=False,bottom=False)
```

```python
sns.distplot(diamonds['carat'],kde=False)
```

```python
tips=sns.load_dataset("tips")
```

```python
sns.set_style('whitegrid')
sns.violinplot(data=tips,x='day',y='total_bill',hue='sex',split=True,
               inner='quartile',palette=['#AECCE4','#FFFCBB'])
sns.despine(top=True,bottom=True,left=True,right=True)
```

```python
ax=sns.barplot(data=titanic,x='class',y='survived',hue='sex')
sns.despine(top=True,bottom=True,left=True,right=True)
plt.ylabel("survival probability")
plt.legend(title='sex',loc='center right',bbox_to_anchor=(1.2,0.5),
           frameon=False)
```

```python
fmri=sns.load_dataset('fmri')
sns.set_style('darkgrid')
sns.lineplot(data=fmri,x='timepoint',y='signal',hue='region',style='event')
plt.legend()
plt.show()
```

```python
iris=sns.load_dataset("iris")
```

```python
sns.set_style('white')
sns.scatterplot(data=iris,x='sepal_length',y='sepal_width',hue='species')
sns.despine(top=True,right=True)
plt.xticks(np.arange(5,9))
plt.legend(loc='center right',bbox_to_anchor=(1.2,0.43),frameon=False)
```

```python
sns.set_style('white')
for specie in iris['species'].unique():
    sns.kdeplot(data=iris[iris['species']==specie],x='petal_length',
                label=specie)
sns.despine(top=True,right=True)
plt.legend(title='species',loc='center right',bbox_to_anchor=(1.24,0.51),
           frameon=False)
```

```python
sns.jointplot(data=iris,x='sepal_length',y='sepal_width',
              marginal_kws={'color': '#9494c6'})
```

```python
sns.pairplot(data=iris,hue='species')
```
